```python
import pandas as pd

data ={
  "email":[
         "win money now",
         "Hello how are you",
         "claim your free prize",
         "meeting at spam",
         "congratulation you won the lottery",
          "let's catch up today",
          "free entry in context",
          "project discussion today"
    ],
    "label":[1,0,1,0,1,0,1,0]

}

df = pd.DataFrame(data)

print(data)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
x = vectorizer.fit_transform(df["email"])
y = df["label"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)

```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier()
gb.fit(x_train,y_train)
```

```python
test_email = ["claim your free prize"]

test_vector = vectorizer.transform(test_email)

def explain(pred):
    return "Spam" if pred == 1 else "Not Spam"

print("DT:", explain(dt.predict(test_vector)[0]))
print("RF:", explain(rf.predict(test_vector)[0]))
print("GB:", explain(gb.predict(test_vector)[0]))
```
